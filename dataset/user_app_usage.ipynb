```python
import csv
import numpy as np
import os
import pandas as pd
```

```python
user_app_usage = pd.read_csv('user_app_usage.csv',header=None,chunksize=1024)
user_app_usage.columns = ['uId','appId','duration','times','use_date']
```

```python
def change_user_app_usage(each):
    def f(df):
        duration = df['duration']
        app = df['appId']
        time = df['times']
        data = {}
        for i,j,k in zip(app,time,duration):
            data[i]=[j,k]
    #     df = pd.Series(data)
        return data
    temp = each.groupby(['uId','use_date']).apply(f)
    temp.index.names = ['','']
    df_temp = temp.unstack()
    return df_temp
```

```python
for each in user_app_usage:
    each.columns = ['uId','appId','duration','times','use_date']
    new_each = change_user_app_usage(each)
    print(len(each))
    print(type(each))
    break
```
